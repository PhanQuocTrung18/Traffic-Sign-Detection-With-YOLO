# Visualize data before augment data 

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r'C:\Users\LAPTOP\Desktop\Traffic Sign Detection With Yolo\Result data\data(-80%).csv'
data = pd.read_csv(file_path)
class_counts = data['class'].value_counts()
single_color = 'skyblue' 
plt.figure(figsize=(10, 6))
bars = plt.bar(class_counts.index, class_counts.values, color=single_color)
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), va='bottom', ha='center')
plt.xlabel('Class')
plt.ylabel('Number of Images')
plt.title('Distribution of Classes')
plt.xticks(class_counts.index)
plt.show()

```
